```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel files
smap_df = pd.read_csv('SMAP.csv')
psm_df = pd.read_csv('PSM.csv')
msl_df = pd.read_csv('MSL.csv')

# Extract F1 scores (column E, index 4), rows 1 to 6
smap_f1 = smap_df.iloc[:5, 4].values * 100  # Convert to percentage
psm_f1 = psm_df.iloc[:5, 4].values * 100
msl_f1 = msl_df.iloc[:5, 4].values * 100

# X-axis labels for Multi-scale Patch-sizes
x_labels = ['[1]', '[3]', '[5]', '[1,3]', '[1,5]', '[3,5]', '[1,3,5]'][:5]
x = range(len(x_labels))

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(x, msl_f1, marker='s', color='blue', label='MSL', linewidth=2)
plt.plot(x, smap_f1, marker='^', color='green', label='SMAP', linewidth=2)
plt.plot(x, psm_f1, marker='o', color='red', label='PSM', linewidth=2)

plt.xticks(x, x_labels, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Multi-scale Patch-sizes', fontsize=12)
plt.ylabel('F1 Score %', fontsize=12)
plt.title('Multi-scale size', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=9)
plt.tight_layout()
plt.savefig("f1_score_plot.png", dpi=200)  # Save with 200 DPI
plt.show()
```
